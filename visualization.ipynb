```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
train = pd.read_csv("./train.csv", encoding="utf-8", index_col=0)
test = pd.read_csv("./test.csv", encoding="utf-8", index_col=0)
submission = pd.read_csv("./sample_submission.csv", encoding="utf-8", index_col=0)
```

```python
sum(sum(train.isnull().values))
```

```python
choice = 100
choice_id = list(np.random.choice(train.index, choice, replace=False))
```

```python
plt.figure(figsize=(20,8))

for i in choice_id:
    plt.plot(train.columns, train.loc[i].values)
    plt.xticks(rotation='60');
```

```python
plt.figure(figsize=(20,8))

plt.subplot(311)
plt.scatter(range(len(train)), np.mean(train.values, axis=1), s=1, c="C0", alpha=0.7, label="mean")
plt.legend()

plt.subplot(312)
plt.scatter(range(len(train)), np.median(train.values, axis=1), s=1, c="C1", alpha=0.7, label="median")
plt.legend()

plt.subplot(313)
plt.scatter(range(len(train)), np.max(train.values, axis=1), s=1, c="C2", alpha=0.7, label="max")
plt.legend();
```
